```python
# Import necessary libraries
import pandas as pd
```

```python
# Access Concrete Dataset

```

```python
# Turn the data into clusters, ie. Make Blobs
from sklearn.datasets import make_blobs

X,y = make_blobs(n_samples=1000, centers=3, cluster_std=0.60, random_state=42, n_features=9)
X = pd.DataFrame(X, columns=['Feature1','Feature2','Feature3','Feature4','Feature5','Feature6','Feature7','Feature8','Feature9'])
```

```python
# perform PCA on the data
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
data_pca = pca.fit_transform(X)
data_pca = pd.DataFrame(data_pca, columns=['PC1','PC2'])
```

```python
# Plot PCA data

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(data_pca['PC1'], data_pca['PC2'],s=20)
plt.xlabel('Principal Component 1')
plt.ylabel('Principle Component 2')
plt.title('PCA of Blobs Data')
plt.show()
```

```python
# Perform KMeans Clustering on the data
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42)
model.fit(data_pca)
labels = model.labels_
centers = model.cluster_centers_
```

```python
# Plot data after performing KMeans
plt.figure(figsize = (10,6))
plt.scatter(data_pca['PC1'], data_pca['PC2'], c=labels, s=20, cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.75, marker='X',label='Centroids')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans Clustering on PCA Data')
plt.legend()
```

```python
# Evaluate model using silhouette score
from sklearn.metrics import silhouette_score
score=silhouette_score(X, labels)
print(f"Silhouette Score: {score:.2f}")
```
